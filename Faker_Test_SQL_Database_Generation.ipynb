```python
from faker import Faker
from sqlalchemy import create_engine, Column, Numeric, Integer, String, DateTime, ForeignKey
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
import pandas as pd
import random


# Configure SQLAlchemy
engine = create_engine('sqlite:///customers.db')  # Adjust connection string as needed
Base = declarative_base()

# Define Customer model
class Customer(Base):
    __tablename__ = 'customers'

    customer_id = Column(Integer, primary_key=True)
    first_name = Column(String(255))
    last_name = Column(String(255))
    email = Column(String(255))
    phone_number = Column(String(20))
    address_line_1 = Column(String(255))
    address_line_2 = Column(String(255))
    city = Column(String(255))
    state = Column(String(50))
    postal_code = Column(String(20))
    country = Column(String(50))

class Product(Base):
    __tablename__ = "products"

    product_id = Column(Integer, primary_key=True)
    product_name = Column(String(255))
    description = Column(String(255))
    category = Column(String(255))
    price = Column(Numeric(10, 2))
    stock_level = Column(Integer)
    image_url = Column(String(255))


class Order(Base):
    __tablename__ = "orders"

    order_id = Column(Integer, primary_key=True)
    customer_id = Column(Integer, ForeignKey("customers.customer_id"))
    order_date = Column(DateTime)
    order_status = Column(String(50))
    total_amount = Column(Numeric(10, 2))


class Order_Items(Base):
    __tablename__ = "order_items"

    order_item_id = Column(Integer, primary_key=True)
    order_id = Column(Integer, ForeignKey("orders.order_id"))
    product_id = Column(Integer, ForeignKey("products.product_id"))
    quantity = Column(Integer)
    unit_price = Column(Numeric(10, 2))


class Payments(Base):
    __tablename__ = "payments"

    payment_id = Column(Integer, primary_key=True)
    order_id = Column(Integer, ForeignKey("orders.order_id"))
    payment_method = Column(String(50))
    payment_date = Column(DateTime)
    transaction_id = Column(String(255))


class Reviews(Base):
    __tablename__ = "reviews"

    review_id = Column(Integer, primary_key=True)
    customer_id = Column(Integer, ForeignKey("customers.customer_id"))
    product_id = Column(Integer, ForeignKey("products.product_id"))
    rating = Column(Integer)
    review_text = Column(String(255))
    review_date = Column(DateTime)


# Create tables (drop existing if needed)
# Base.metadata.reflect(engine)
# if "customers" in Base.metadata.tables:
#     Base.metadata.tables["customers"].drop(engine)
# Base.metadata.create_all(engine)

Base.metadata.drop_all(engine)  # Drop all tables before creating them
Base.metadata.create_all(engine)

# Generate fake data
fake = Faker()
session = sessionmaker(bind=engine)()

for _ in range(10):
    customer = Customer(
        first_name=fake.first_name(),
        last_name=fake.last_name(),
        email=fake.email(),
        phone_number=fake.phone_number(),
        address_line_1=fake.street_address(),
        address_line_2=fake.secondary_address(),
        city=fake.city(),
        state=fake.state_abbr(),
        postal_code=fake.postcode(),
        country=fake.country(),
    )
    session.add(customer)

session.commit()

# Generate and insert fake data for Products
fake = Faker()
categories = ["Electronics", "Clothing", "Home & Garden", "Beauty", "Books"]

for _ in range(20):  # Generate 20 products
    product = Product(
        product_name=fake.word(),
        description=fake.text(),
        category=random.choice(categories),
        price=round(fake.pydecimal(min_value=10, max_value=100, positive=True), 2),
        stock_level=random.randint(10, 100),
        image_url=fake.image_url(),
    )
    session.add(product)

session.commit()



# Query data to pandas dataframe
session = sessionmaker(bind=engine)()
customers_df = pd.read_sql_table("customers", engine)
products_df = pd.read_sql_table("products", engine)
```
